```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(20)
y = np.arange(20)
X,Y = np.meshgrid(x,y)
```

```python
ZX = np.sin(X*0.25)
ZY = np.cos(Y*0.2)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize = (16,8), frameon = False)

ax1.plot(X,Y, "ok")
ax1.plot(X+ZX,Y+ZY, ".b")
ax2.quiver(X,Y,ZX,ZY)
plt.show()
```
